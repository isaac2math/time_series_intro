# Single exponential forecast

```stata
tssmooth exponential sm1=unemp, forecast(60)
```

# plot the result

```stata
line sm1 unemp month, title("Single exponential forecast")
```

---

# Holt Winters exponential smoothing

```stata
tssmooth shwinters hwy=unemp, forecast(60)
```

```stata
line unemp hwy month, title("Holt Winters exponential smoothing")
```
