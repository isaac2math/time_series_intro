# load your data as a table and set time index

```matlab
T      = readtable('AuQuGDP.xlsx');
T.Time = datetime(T.Time);
TT     = table2timetable(T);
```

```matlab
head(TT)
```

# line plot, acf plot and pacf plot of AuGDP

```matlab
figure

plot(TT.Time, TT.AuGDP)

title('line plot of AuGDP')
```

```matlab
[acf,lags,bounds,h]  = autocorr(TT.AuGDP)
```

```matlab
[pacf,lags,bounds,h] = parcorr(TT.AuGDP)
```

# ADF test of AuGDP

```matlab
[h,pValue,stat,cValue,reg] = adftest(TT.AuGDP,'model','TS','lags',0:3)
```

# first order difference your time series and save it as GDP_D1

```matlab
TT.GDP_D1 = [NaN; diff(TT.AuGDP)];
head(TT)
```

# line plot, acf plot and pacf plot of GDP_D1

```matlab
figure

plot(TT.Time, TT.GDP_D1)

title('line plot of GDP_D1')
```

```matlab
[acf,lags,bounds,h]  = autocorr(TT.GDP_D1)
```

```matlab
[pacf,lags,bounds,h] = parcorr(TT.GDP_D1)
```

# ADF test of GDP_D1

```matlab
[h,pValue,stat,cValue,reg] = adftest(TT.GDP_D1,'model','TS','lags',0:3)
```

# MA and AR estimation

```matlab
model_AR5        = arima(5,0,0);
model_AR5_fit    = estimate(model_AR5,TT.GDP_D1,'Display','off'); 
model_AR5_result = summarize(model_AR5_fit)
AR5_AIC          = model_AR5_result.AIC;
AR5_BIC          = model_AR5_result.BIC;
summarize(model_AR5_fit)
```

```matlab
model_MA6        = arima(0,0,6);
model_MA6_fit    = estimate(model_MA6,TT.GDP_D1,'Display','off'); 
model_MA6_result = summarize(model_MA6_fit)
MA6_AIC          = model_MA6_result.AIC;
MA6_BIC          = model_MA6_result.BIC;
summarize(model_MA6_fit)
```
